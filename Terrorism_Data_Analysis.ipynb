```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df = pd.read_csv("Terrorism_Data.csv", encoding='ISO-8859-1')
```

```python

# Check unique values in each column
df.nunique()
```

```python
# Combine day, month, and year columns into a single string column representing date
df['date'] = df.apply(lambda row: pd.to_datetime(f"{row['iday']}-{row['imonth']}-{row['iyear']}", format='%d-%m-%Y', dayfirst=True, errors='coerce'), axis=1)

# Drop the original day, month, and year columns
df.drop(['iday', 'imonth', 'iyear'], axis=1, inplace=True)
```

```python
# Handle missing values
df.dropna(subset=['latitude', 'longitude'], inplace=True) # Remove rows with missing lat/lon values

# Convert date column to datetime data type
df['date'] = pd.to_datetime(df['date'])

# Remove irrelevant columns
df.drop(['approxdate', 'resolution', 'location', 'summary', 'provstate', 'corp1', 'corp2', 'corp3', 'addnotes', 'scite1', 'scite2', 'scite3', 'dbsource', 'INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY'], axis=1, inplace=True)
```

# hot zone of terrorism

```python

# Task 1: Identify countries with the highest number of terrorist incidents
country_counts = df['country_txt'].value_counts().head(10)

# Set up plot figure and axes
fig, ax = plt.subplots(figsize=(10, 7))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='muted', ax=ax)

# Add labels and title
ax.set_title('Top 10 Countries with Highest Number of Terrorist Incidents', fontsize=16, fontweight='bold')
ax.set_xlabel('Country', fontsize=12)
ax.set_ylabel('Number of Incidents', fontsize=12)

# Customize tick labels
ax.tick_params(axis='x', rotation=45, labelsize=10)
ax.tick_params(axis='y', labelsize=10)

# Show plot
plt.show()
```

# security issues and insights derived by EDA

```python
# Task 2: Analyze trends in terrorist attacks over time
df['year'] = pd.to_datetime(df['date']).dt.year
yearly_counts = df['year'].value_counts().sort_index()

# Set up plot figure and axes
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(yearly_counts.index, yearly_counts.values, linewidth=2, color='b', marker='o', markersize=5)

# Add labels and title
ax.set_title('Terrorist Incidents by Year', fontsize=16, fontweight='bold')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Number of Incidents', fontsize=12)

# Set up grid lines
ax.grid(True, linestyle='--', linewidth=0.5, color='gray')

# Customize tick labels
ax.tick_params(axis='x', rotation=45, labelsize=10)
ax.tick_params(axis='y', labelsize=10)

# Show plot
plt.show()

# Plot a heatmap of terrorist attacks by month and year
month_year = df.groupby([df['date'].dt.year, df['date'].dt.month])['eventid'].count().unstack()
plt.figure(figsize=(12,6))
sns.heatmap(month_year, cmap='coolwarm', annot=True, fmt='d')
plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Terrorist Attacks by Month and Year')
plt.show()

plt.figure(figsize=(8,8))
df['attacktype1_txt'].value_counts().plot(kind='pie', startangle=90)
plt.title('Terrorist Attacks by Attack Type')
plt.ylabel('')
plt.legend(bbox_to_anchor=(1,1), loc="upper left", fontsize=10)
plt.tight_layout()
plt.show()

# Task 3: Investigate types of attacks and weapons used
attack_counts = df['attacktype1_txt'].value_counts()
weapon_counts = df['weaptype1_txt'].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
attack_counts.plot(kind='pie', ax=axes[0], autopct='%1.1f%%', startangle=90, legend=False)
weapon_counts.plot(kind='bar', ax=axes[1])
axes[0].set_title('Types of Attacks')
axes[1].set_title('Types of Weapons Used')
plt.show()

# Task 4: Examine targets and motives of terrorist attacks
target_counts = df['targtype1_txt'].value_counts().head(10)
motive_counts = df['motive'].value_counts().head(10)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
target_counts.plot(kind='barh', ax=axes[0])
motive_counts.plot(kind='barh', ax=axes[1])
axes[0].set_title('Top 10 Target Types')
axes[1].set_title('Top 10 Motives')

# Set custom font size for labels
plt.xticks(fontsize=10)

# Rotate labels in the first subplot
axes[0].set_yticklabels(target_counts.index, rotation=0, ha='right')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()

# Task 5: Analyze groups or organizations responsible for attacks
group_counts = df['gname'].value_counts().head(10)
colors = plt.cm.tab20c(np.arange(len(group_counts)))  # Choose custom colors
fig, ax = plt.subplots(figsize=(10, 6))
group_counts.plot(kind='bar', color=colors, ax=ax)

# Add title and labels with custom font size and style
ax.set_title('Top 10 Terrorist Groups', fontsize=16, fontweight='bold')
ax.set_xlabel('Group/Organization', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Incidents', fontsize=12, fontweight='bold')

# Set custom font size for tick labels
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

# Add grid lines
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()

# Task 6: Identify patterns or correlations between variables
# Example: Scatter plot between number of killed and wounded
plt.scatter(df['nkill'], df['nwound'], alpha=0.5, c='red', edgecolors='black', s=50)

# Add title and labels with custom font size and style
plt.title('Number of Killed vs. Number of Wounded', fontsize=16, fontweight='bold')
plt.xlabel('Number of Killed', fontsize=12, fontweight='bold')
plt.ylabel('Number of Wounded', fontsize=12, fontweight='bold')

# Set custom font size for tick labels
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)

# Add grid lines
plt.grid(linestyle='--', alpha=0.5)

# Customize marker properties
marker = plt.Line2D([0, 0], [0, 0], color='red', marker='o', linestyle='', markersize=8, markerfacecolor='white', markeredgecolor='black', alpha=0.5)
plt.legend([marker], ['Incident'], loc='upper right', fontsize=10)

plt.show()


# Task 7: Explore impact of terrorism on different sectors or industries
# Example: Bar chart of number of attacks on different sectors
sector_counts = df['targtype2_txt'].value_counts().head(10)

# Create a horizontal bar chart with custom colors, edge color, and width
bar_colors = ['#0099ff', '#ff6666', '#66ff66', '#ffcc99', '#ff99ff', '#9999ff', '#ffcc66', '#99ff99', '#cc99ff', '#ff9999']
bar_edge_color = 'black'
bar_width = 0.7
sector_counts.plot(kind='barh', color=bar_colors, edgecolor=bar_edge_color, figsize=(10, 6), width=bar_width)

# Add title and labels with custom font size and style
plt.title('Top 10 Sectors with Highest Number of Attacks', fontsize=16, fontweight='bold')
plt.xlabel('Number of Incidents', fontsize=12, fontweight='bold')
plt.ylabel('Sector', fontsize=12, fontweight='bold')

# Set custom font size for tick labels
plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=10)

# Add grid lines
plt.grid(linestyle='--', alpha=0.5)

plt.show()


```
